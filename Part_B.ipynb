# Part B Grouped Analysis
Dataset: seaborn tips dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
df.head()
```

## 1. Perform grouped analysis
Group by categorical column and compute mean

```python
grouped = df.groupby('day')[['total_bill','tip']].mean()
grouped
```

## 2. Visualize grouped results using bar plot

```python
grouped.plot(kind='bar')
plt.title('Mean Total Bill and Tip by Day')
plt.show()
```

## 3. Compare two numerical features using line chart and KDE plot

```python
df[['total_bill','tip']].plot(kind='line')
plt.title('Line Comparison')
plt.show()
```

```python
df['total_bill'].plot(kind='kde')
df['tip'].plot(kind='kde')
plt.title('KDE Comparison')
plt.show()
```

## 4. Insights
Higher bills and tips are observed on weekends. Tips follow total bill trend. KDE shows both distributions are right skewed.
