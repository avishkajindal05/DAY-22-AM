# Part A Concept Application
Dataset: seaborn tips dataset loaded from library

## 1. Data Selection using Pandas
Use loc and iloc with at least 3 examples each

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('tips')
df.head()
```

```python
# loc examples
df.loc[0:5, ['total_bill', 'tip']]
df.loc[df['sex'] == 'Male', ['total_bill', 'tip']].head()
df.loc[(df['day'] == 'Sun') & (df['time'] == 'Dinner'), ['total_bill', 'tip']].head()
```

```python
# iloc examples
df.iloc[0:5, 0:3]
df.iloc[10:20, [0, 2, 4]]
df.iloc[:, [0, 1]].head()
```

## 2. Filtering Data
Filter rows based on multiple conditions and extract subsets

```python
filtered = df[(df['total_bill'] > 20) & (df['tip'] > 3)]
filtered.head()
```

```python
subset = df[df['day'] == 'Sun']
subset.head()
```

## 3. Descriptive Statistics
Use describe and interpret mean, std, min, max

```python
stats = df.describe()
stats
```

```python
mean_val = df['total_bill'].mean()
std_val = df['total_bill'].std()
min_val = df['total_bill'].min()
max_val = df['total_bill'].max()

mean_val, std_val, min_val, max_val
```

Mean gives average bill value. Std shows spread. Min and max show range.

## 4. Histogram
Plot histogram and interpret distribution

```python
import matplotlib.pyplot as plt

df['total_bill'].hist(bins=20)
plt.title('Histogram of Total Bill')
plt.show()
```

Distribution is slightly right skewed with most values in lower range.

## 5. Bar Plot
Categorical vs numerical comparison

```python
df.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.title('Average Total Bill by Day')
plt.show()
```

Weekend days tend to have higher average bills.

## 6. Line Chart
Trend over index

```python
df['total_bill'].plot(kind='line')
plt.title('Total Bill Trend')
plt.show()
```

Shows variation across observations.

## 7. KDE Plot
Density distribution and comparison with histogram

```python
df['total_bill'].plot(kind='kde')
plt.title('KDE Plot of Total Bill')
plt.show()
```

KDE smooths the histogram and highlights density peaks.
