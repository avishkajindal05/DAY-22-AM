# Data Analysis with Pandas and Visualization with Matplotlib

## Introduction
This notebook demonstrates how to perform data analysis using Pandas and visualize data using Matplotlib.
We use a built-in dataset to ensure easy reproducibility.

## Load Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
df.head()
```

## Basic Data Exploration

```python
df.shape
df.columns
df.info()
```

## Descriptive Statistics

```python
df.describe()
```

## Data Selection

```python
df[['total_bill','tip']].head()
df.loc[df['day']=='Sun', ['total_bill','tip']].head()
df.iloc[0:5, 0:3]
```

## Filtering Data

```python
df[(df['total_bill']>20) & (df['tip']>3)].head()
```

## Grouped Analysis

```python
df.groupby('day')[['total_bill','tip']].mean()
```

## Histogram

```python
df['total_bill'].hist(bins=20)
plt.title('Total Bill Distribution')
plt.show()
```

## Bar Plot

```python
df.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.title('Average Bill by Day')
plt.show()
```

## Line Chart

```python
df['total_bill'].plot(kind='line')
plt.title('Total Bill Trend')
plt.show()
```

## KDE Plot

```python
df['total_bill'].plot(kind='kde')
plt.title('Density Plot')
plt.show()
```
